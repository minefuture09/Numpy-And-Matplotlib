```python
import matplotlib.pyplot as plt 
import csv
x = []
y = []

with open("sample.txt","r") as f:
    plots = csv.reader(f)
    for col in plots:
        x.append(col[0])
        y.append(col[1])
        
plt.plot(x,y,label="file")
plt.title('File Handling')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```
